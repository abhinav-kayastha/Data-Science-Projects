```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

# getting the path to the training data
MobileTrainPriceDataPath = "train.xls"

# making pandas read the data
MobileTrainPriceData = pd.read_csv(MobileTrainPriceDataPath)

# drops a row of data if there are any variables that were not recorded
MobileTrainPriceData.dropna(axis=0)

# gives the summarized shape of the training data set 
MobileTrainPriceData.describe()
```

```python
# selecting the prediction target
y_train = MobileTrainPriceData.price_range

# choosing the features that will be used in order to make the prediction
MobileFeatures = ['battery_power', 'clock_speed', 'int_memory', 'wifi', 'four_g']

# getting just the features and assigning them to variable X
X_train = MobileTrainPriceData[MobileFeatures]

# geting the first few data
X_train.head()
```

```python
# making a decision tree regression object, random_state=1 means to have some randomness in model training and using the number ensures same
# results in each run  
MobilePriceRangeModel = DecisionTreeRegressor(random_state=1)

# fitting the model with the data
MobilePriceRangeModel.fit(X_train, y_train)
```
